## Pyplot API vs Object Oriented (OO) API
* Pylot Api -> Quickly
* OO Api -> Advanced

### Pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed");
plt.plot(x, np.cos(x), color="red", label="cos(x)")
plt.xlabel("x")
plt.ylabel("Sin(x)")
# plt.xlim([0, 4]) # Limit range x
# plt.ylim([0, 1]) # Limit range y

plt.title("Sin(x) and Cos(x)") # title

plt.axis()

plt.legend() # Space Line in plot
```

### Object-Oriented API

```python
# Step 1: Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Step 2: Set up plot
figure, ax = plt.subplots(figsize=(5, 5))

# Step 3: 
ax.plot(x, y);

# Set figure
ax.set(title="A simple plot",
       xlabel="x",
       ylabel="y"
      )
```

## Common Type in Matplotlib Plots
* Line
* Scatter
* Bar
* Hist
* Subplot

### Line

```python
# Create an array
a = np.linspace(0, 10, 100)
a[:5]
```

```python
# OO API create a line chart 

fig, ax = plt.subplots()
ax.plot(a, a**3)
```

### Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x)); # y = e^x
```

```python
# OO API
# Prepare new data
rng = np.random.RandomState(0)

x = rng.randn(10)
y = rng.randn(10)
colors = rng.rand(100) # Random color
sizes = 1000*rng.rand(100)
fig, ax = plt.subplots()

img1 = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.5);
fig.colorbar(img1);
```

### Bar
* Vertical
* Horizontal

```python
# Prepare data

soft_drink_prices = {
    "Coke": 10,
    "Pepsi": 12,
    "Sprite": 8
}

fig, ax = plt.subplots()

ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title="Soft Drink - Price", ylabel="Price($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
```
