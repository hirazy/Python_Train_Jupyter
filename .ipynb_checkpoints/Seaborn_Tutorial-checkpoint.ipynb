* **Box Plot**: Phương pháp để mô tả đồ thị các nhóm dữ liệu số thông qua các phần tử của chúng.
* **Join Plot**
* **Pair Plot**

### Distribution Plot
* **Univariate Distribution of Observation**

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
sns.histplot(data=tips_df["total_bill"]);
```

```python
tips_df["total_bill"].value_counts().sort_values(ascending=False);
```

# KDE Plot
* A kernel density estimate(KDE) plots is a method for visualizing the distribution of observations in a dataset, analagous to a histogram.
* KDE represents the data using a continuous probability density curve in one or more dimensions

```python
sns.kdeplot(data=tips_df["total_bill"]);
```

### Displot

```python
sns.displot(data=tips_df, x="total_bill", col="time",kde=True);
```

### Bar Plot

```python
tips_df.head()
```

```python
sns.barplot(data=tips_df, x="sex", y="tip", estimator=np.mean);
```

### Box Plot
* Box Plot is good indication of how the value in the data are spred out 

```python
sns.boxplot(data=tips_df, x="day", y="total_bill", hue="sex", palette="Blues");

plt.legend(loc=0);
```

### Facet Grid

```python
tips_df.head()
```

```python
tips_fg = sns.FacetGrid(data=tips_df, row="smoker", col="time");
tips_fg.map(sns.scatterplot, 'total_bill', 'tip');
```

```python
new_fg = sns.FacetGrid(data=tips_df, col="sex", 
                       hue="smoker", # Different Color
                       palette='Set2',
                       col_order=["Female", "Male"], 
                        aspect=1.4);
new_fg.map(sns.scatterplot, 'total_bill', 'tip', s=75, alpha=.6);
new_fg.add_legend();
```

### Join Plot
* joint distribution between two variable along with each variable's marginal distribution
